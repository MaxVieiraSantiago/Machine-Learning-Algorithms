```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

%pylab inline
```

```python
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Machine-Learning-Algorithms/master/Dataset/Iris.csv'

dados = pd.read_csv( url )
dados.head()
```

```python
X = dados.drop( ['Id', 'Species' ], axis=1 )
y = dados['Species']
```

```python
X_train, X_teste, y_train, y_teste = train_test_split( X, y, test_size=0.30, random_state=1 )
```

```python
modelo = DecisionTreeClassifier()
y_pred = modelo.fit( X_train, y_train ).predict( X_teste )
```

```python
pylab.figure(figsize=(25,15))

plot_tree( modelo, feature_names=X_train.columns )
```
