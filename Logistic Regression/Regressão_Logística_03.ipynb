```python
import numpy as np
import matplotlib.pyplot as plt
import math

X = np.linspace( -10, 10, 100 )
y = 1 / ( 1 + np.exp( -X ) )

plt.figure( figsize=(10,5))
plt.plot( X, y, linewidth=3 )
plt.xlabel( 'X', fontsize=15 )
plt.ylabel( 'h(X)', fontsize=15 )
plt.savefig( 'logistic-function.eps' )
plt.show()
```
