```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Machine-Learning-Algorithms/master/Dataset/wineQualityWhites.csv'
dados_white = pd.read_csv( url )
dados_white.head()
```

```python
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Machine-Learning-Algorithms/master/Dataset/wineQualityReds.csv'
dados_red = pd.read_csv( url )
dados_red.head()
```

```python
dados = pd.concat( [ dados_red, dados_white ] )
dados.head()
```

```python
dados.info()
```

```python
X = dados.drop( [ 'quality', 'Type' ], axis=1 )
y = dados['quality']
```

```python
X_train, X_teste, y_train, y_teste = train_test_split( X, y, test_size=0.20, random_state = 0 )
```

```python
modelo = RandomForestClassifier( n_estimators=600, min_samples_leaf=1, n_jobs=-1 )

modelo.fit( X_train, y_train );
```

```python
y_pred = modelo.predict( X_teste )

mse = np.sqrt( mean_squared_error( y_teste, y_pred ) ) * 100
accuracy = accuracy_score( y_teste, y_pred ) * 100

print(f'Erro médio quadrado = {mse:,.2f}%' )
print(f'Acurácia do modelo  = {accuracy:,.2f}%' )
```
